```python
#Example Problem 3.3 Construct IPR of a vertical well in a saturated oil reservoir using Vogel’s equation. The following data are given:

# Porosity: φ = 0.19
# Effective horizontal permeability: k = 8.2 md
# Pay zone thickness: h = 53 ft
# Reservoir pressure: rp = 5 651 psia
# Bubble-point pressure: pb = 5651 psia
# Fluid formation volume factor: Bo = 1.1
# Fluid viscosity: mo = 1.7 cp
# Total compressibility: ct = 0.0000129 psi21
# Drainage area: A = 640 acres (re = 2980 ft)
# Wellbore radius: rw = 0.328 ft
# Skin factor: S = 0
```

```python
#so first we calculate the productivity index j

def productivity_index(k, h, B,mo, re, rw, S ):
  j=[(k*h)/((141.2*B*mo) * (np.log(re/rw))-(3/4)+S)]

  return f'{j} is the productivity index'
```

```python
productivity_index(8.2, 53, 1.1, 1.7, 2980, 0.328, 0)
```
